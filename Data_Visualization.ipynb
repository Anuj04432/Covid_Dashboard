```python
sns.set_theme(style="whitegrid")

plt.rcParams.update({
    "figure.facecolor": "#f4f6f9",
    "axes.facecolor": "#ffffff",
    "axes.edgecolor": "#dddddd",
    "axes.titleweight": "bold",
    "axes.titlesize": 11,
    "axes.labelsize": 9,
    "xtick.labelsize": 8,
    "ytick.labelsize": 8
})
```

```python
data = pd.read_csv("OneDrive/Documents/datasets/country_wise_latest.csv")
```

```python
data.sort_values(by="Deaths",ascending=False).head()
```

```python
def bar_chart(ax):
    top10 = data.sort_values(by="Confirmed", ascending=False).head(10)

    colors = ["red" if i==0 else "orange" for i in range(len(top10))]
    ax.bar(top10["Country/Region"], top10["Confirmed"], color=colors)

    ax.set_title("USA Dominates COVID Cases Among Top Countries", fontsize=10)
    ax.set_ylabel("Cases")

    ax.tick_params(axis='x', rotation=45)

    # annotation
    ax.annotate("Highest cases",
                xy=(0, top10["Confirmed"].iloc[0]),
                xytext=(1, top10["Confirmed"].iloc[0]*0.7),
                arrowprops=dict(arrowstyle="->"))
```

```python
def pie_chart(ax):
    top5 = data.sort_values(by="Deaths", ascending=False).head(5)

    ax.pie(top5["Deaths"], labels=top5["Country/Region"], autopct='%1.1f%%')

    ax.set_title("Top Countries by Death Share")
```

```python
def box_plot(ax):
    ax.boxplot(
        [data["Confirmed"], data["Deaths"], data["Active"]],
        patch_artist=True,
        boxprops=dict(facecolor='orange')
    )

    ax.set_xticks([1,2,3])
    ax.set_xticklabels(["Confirmed","Deaths","Active"])

    ax.set_title("Spread of Cases, Deaths and Active Cases")
```

```python
def histogram(ax):
    sns.histplot(data["Confirmed"], bins=20, kde=True, ax=ax,color="red")

    ax.set_xscale('log')
    ax.set_title("COVID Cases Distribution is Highly Skewed")
```

```python
def scatter_plot(ax):
    sns.scatterplot(
        x="Confirmed",
        y="Deaths",
        hue="WHO Region",
        data=data,
        alpha=0.7,
        s=60,
        ax=ax
    )

    ax.set_title("Deaths Increase with Confirmed Cases (Positive Correlation)")
```

```python
def heatmap(ax):
    cols = ["Confirmed","Deaths","Recovered","Active"]

    sns.heatmap(
        data[cols].corr(),
        annot=True,
        cmap="coolwarm",
        ax=ax
    )

    ax.set_title("Strong Correlation Between COVID Metrics")
```

```python
def line_chart(ax):
    top10 = data.sort_values(by="Confirmed", ascending=False).head(10)

    ax.plot(top10["Country/Region"], top10["Confirmed"], marker="*")
    ax.plot(top10["Country/Region"], top10["Deaths"], marker="o")
    ax.plot(top10["Country/Region"], top10["Recovered"], marker="x")

    ax.set_title("Recovered Cases Follow Confirmed Case Trends")
    ax.set_ylabel("Cases")

    ax.tick_params(axis='x', rotation=45)
    ax.legend(["Confirmed","Deaths","Recovered"])
```

```python
def count_plot(ax):
    sns.countplot(x="WHO Region", data=data, color="orange", ax=ax)

    ax.tick_params(axis='x', rotation=45)
    ax.set_title("Europe and Africa Have Highest Number of Affected Countries")
```

```python
def donut_chart(ax):
    top5 = data.sort_values(by="Recovered", ascending=False).head(5)

    ax.pie(top5["Recovered"], labels=top5["Country/Region"], autopct='%1.0f%%')

    centre = plt.Circle((0,0),0.5,color='white')
    ax.add_artist(centre)

    ax.set_title("Few Countries Contribute Majority of Recoveries")
```

```python
def barh_chart(ax):
    top10 = data.sort_values(by="Active", ascending=False).head(10)

    ax.barh(top10["Country/Region"], top10["Active"], color="green")

    ax.set_title("Top Countries by Active Cases")
    ax.set_xlabel("Active Cases")

    ax.invert_yaxis()
```

```python
def kde_plot(ax):
    sns.kdeplot(data["New cases"], label="New Cases", ax=ax)
    sns.kdeplot(data["New deaths"], label="New Deaths", ax=ax)

    ax.set_title("Recent Spread of COVID Cases and Deaths")
    ax.legend()
```

```python
def area_chart(ax):
    top10 = data.sort_values(by="Active", ascending=False).head(10)

    x = range(len(top10))

    ax.stackplot(
        x,
        top10["Active"],
        top10["Recovered"],
        top10["Deaths"],
        labels=["Active","Recovered","Deaths"]
    )

    ax.set_xticks(x)
    ax.set_xticklabels(top10["Country/Region"], rotation=45)

    ax.set_title("Active Cases Drive Current COVID Burden")
    ax.legend()
```

```python
def dashboard():
    fig, axs = plt.subplots(3, 2, figsize=(18, 16))

    bar_chart(axs[0,0])
    line_chart(axs[0,1])

    scatter_plot(axs[1,0])
    heatmap(axs[1,1])

    histogram(axs[2,0])
    donut_chart(axs[2,1])

    fig.suptitle("COVID-19 Insight Dashboard", fontsize=16,color="darkred")

    plt.tight_layout()
    plt.show()

dashboard()
```

```python
def dashboard_advanced():
    fig, axs = plt.subplots(3, 2, figsize=(18, 16))

    pie_chart(axs[0,0])
    box_plot(axs[0,1])

    count_plot(axs[1,0])
    area_chart(axs[1,1])

    kde_plot(axs[2,0])
    barh_chart(axs[2,1])

    fig.suptitle("COVID-19 Dashboard - Advanced Insights", fontsize=16)

    plt.tight_layout()
    plt.show()
dashboard_advanced()
```
